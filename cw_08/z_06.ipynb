#Import bibliotek

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
```

#Zadanie 1

```python
r = 0.2
startx = 0
starty = 0
margines = 0.1

fig, ax = plt.subplots()
circle = plt.Circle((startx, starty), r, color='red')
ax.add_patch(circle)

ax.set_xlim(startx-r-margines, startx+r+margines)
ax.set_ylim(starty-r-margines, starty+r+margines)
ax.set_aspect(1)
ax.axis("off")

plt.show()
```

```python
chiny = 1408000000
indie = 1454000000

r_indie = 1
r_chiny = (chiny / indie) ** 0.5 * r_indie

fig, ax = plt.subplots(figsize=(6, 3))
ax.set_aspect('equal')

circle_chiny = Circle((0, 0), r_chiny, color='red')
circle_indie = Circle((2 * r_indie + 0.4, 0), r_indie, color='orange')

ax.add_patch(circle_chiny)
ax.add_patch(circle_indie)

plt.text(0, 0, 'Chiny', color='white', ha='center', va='center', fontsize=12)
plt.text(2 * r_indie + 0.4, 0, 'Indie', color='white', ha='center', va='center', fontsize=12)

percentage = (chiny / indie) * 100
plt.title(f'Udział populacji Chin w populacji Indii: {percentage:.2f}%', fontsize=12)

ax.plot([-3, 4], [r_chiny, r_chiny], color='black', linewidth=1)
ax.plot([-3, 4], [-1 * r_chiny, -1 * r_chiny], color='black', linewidth=1)

ax.axis('off')

ax.set_xlim(-1.5, 3.5)
ax.set_ylim(-1.3, 1.3)

plt.show()
```

# Zadanie 2

```python
rok = ['2010', '2011', '2012', '2013', '2014', '2015', '2016']
wydatki = [258, 363, 1282]
wydatki_lata_zagregowane = 4862
sredni_wydatek_zagregowany = wydatki_lata_zagregowane / 4
for _ in range(4):
    wydatki.append(sredni_wydatek_zagregowany)

kolory = ['blue'] * 3 + ['orange'] * 4

plt.figure(figsize=(10, 6))
bars = plt.bar(rok, wydatki, color=kolory, alpha=0.8)

plt.title("Wydatki na badania i rozwój w Polsce")
plt.ylabel("Wydatki (mln zł)")
plt.xlabel("Rok")
plt.ylim(0, max(wydatki) + 500)

plt.legend(handles=[
    plt.Rectangle((0, 0), 1, 1, color='blue', alpha=0.8),
    plt.Rectangle((0, 0), 1, 1, color='orange', alpha=0.8)
], labels=['Wydatki rzeczywiste', 'Wydatki planowane (oszacowane)'])

plt.tight_layout()
plt.show()
```
